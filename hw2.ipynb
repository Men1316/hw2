```python
import seaborn as sns  # serve per la heatmap
def plot_confusion_matrix_multiclass(cm,class_names):
    """
    Visualizza una matrice di confusione come heatmap.

    Parameters:
    -----------
    cm : numpy.ndarray
        La matrice di confusione da visualizzare
    """
    
    # Visualizza la heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, plot_tree
from tensorflow.keras.datasets import cifar10

class_names = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
]


# Carichiamo il dataset CIFAR10
(X_train_full, y_train_full), (X_test, y_test) = cifar10.load_data()
y_train_full = y_train_full.ravel()
y_test = y_test.ravel()
print(f"Shape train full: {X_train_full.shape}, Shape test: {X_test.shape}")

# Opzione di sotto-campionamento
subset_fraction = 0.2 # possiamo impostare <1.0 per ridurre i tempi
if subset_fraction < 1.0:
    n_samples = int(len(X_train_full) * subset_fraction)
    X_train_full = X_train_full[:n_samples]
    y_train_full = y_train_full[:n_samples]
    print(f"Sotto-campionamento: nuovo train shape = {X_train_full.shape}")

# Dividiamo il dataset in train, validation e test
train_frac = 0.6
val_frac = 0.2

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full,
    train_size=train_frac,
    random_state=42,
    stratify=y_train_full
)

print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")

# Appiattiamo e scaliamo 
X_train_flat = X_train.reshape(len(X_train), -1)
X_val_flat   = X_val.reshape(len(X_val), -1)
X_test_flat  = X_test.reshape(len(X_test), -1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_flat)
X_val_scaled   = scaler.transform(X_val_flat)
X_test_scaled  = scaler.transform(X_test_flat)
#Applichiamo PCA
pca = PCA(n_components=0.95, random_state=42)   # conserva il 95% della varianza
X_train_scaled = pca.fit_transform(X_train_scaled)
X_val_scaled   = pca.transform(X_val_scaled)
X_test_scaled  = pca.transform(X_test_scaled)
print("Numero di componenti PCA:", pca.n_components_)
print(f"Dopo PCA: train.shape = {X_train_scaled.shape}")



# Valutiamo il modello
def evaluate_model(name, model, X_tr, y_tr, X_v, y_v, X_te, y_te):
    #accuracy, confusion matrix e classification report.
    print(f"\n--- {name} ---")
    # Training
    model.fit(X_tr, y_tr)
    # Predizioni
    pred_val = model.predict(X_v)
    pred_test = model.predict(X_te)
    # Metriche
    acc_val = accuracy_score(y_v, pred_val)
    acc_test = accuracy_score(y_te, pred_test)
    print(f"Accuracy validation: {acc_val:.4f}")
    print(f"Accuracy test: {acc_test:.4f}")
    print("Confusion Matrix (test):")
    cm = confusion_matrix(y_te, pred_test)
    plot_confusion_matrix_multiclass(cm,class_names)
    
    print("Classification Report (test):")
    print(classification_report(y_te, pred_test,target_names=class_names))

# Regressione Logistica
logreg = LogisticRegression(max_iter=1000, solver='lbfgs', multi_class='auto', random_state=42)
evaluate_model('Logistic Regression', logreg, X_train_scaled, y_train, X_val_scaled, y_val, X_test_scaled, y_test)

# k-NN
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
evaluate_model('k-NN', knn, X_train_scaled, y_train, X_val_scaled, y_val, X_test_scaled, y_test)

# SVM
svm = SVC(C=0.01, kernel='linear', decision_function_shape='ovr')
evaluate_model('SVM (linear)', svm, X_train_scaled, y_train, X_val_scaled, y_val, X_test_scaled, y_test)

# Decision Tree
tree = DecisionTreeClassifier(criterion='entropy', max_depth=None, random_state=42)
evaluate_model('Decision Tree', tree, X_train_scaled, y_train, X_val_scaled, y_val, X_test_scaled, y_test)

# Model Selection e Gridsearch
param_grids = {
    'LogisticRegression': {'C': [0.1, 1.0, 10.0]},
    'KNN': {'n_neighbors': [3, 5, 7]},
    'SVM': {'C': [0.1, 1.0], 'kernel': ['linear', 'rbf']},
    'DecisionTree': {'max_depth': [10, 20, None], 'min_samples_leaf': [1, 5]}
}

models = {
    'LogisticRegression': LogisticRegression(max_iter=1000, solver='lbfgs', multi_class='auto', random_state=42),
    'KNN': KNeighborsClassifier(metric='euclidean'),
    'SVM': SVC(random_state=42),
    'DecisionTree': DecisionTreeClassifier(criterion='entropy', random_state=42)
}

for name in models:
    print(f"\n*** Grid Search per {name} ***")
    gs = GridSearchCV(models[name], param_grids[name], cv=3, scoring='accuracy', n_jobs=-1)
    gs.fit(X_train_scaled, y_train)
    print(f"Best params: {gs.best_params_}")
    print(f"Best CV accuracy: {gs.best_score_:.4f}")
    best = gs.best_estimator_
    evaluate_model(f"{name} (best)", best, X_train_scaled, y_train, X_val_scaled, y_val, X_test_scaled, y_test)
```
